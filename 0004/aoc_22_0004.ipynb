```python
filepath = '../../../Advent_Of_Code_2022/DataFiles/data_day_0002.txt'

with open(filepath, 'r') as fp:

    data = fp.read()
```

```python
data = data.split('\n')

for i in range(len(data)):

    data[i] = data[i].split(' ')

data
```

```python
player_one = {
    'A': ['Rock', 1],
    'B': ['Paper', 2],
    'C': ['Scissors', 3]
}

player_two = {
    'X': 0,
    'Y': 3,
    'Z': 6
}

scores = list()

temp_data = [['A', 'X'], ['B', 'X'], ['C', 'X'], ['A', 'Y'], ['B', 'Y'], ['C', 'Y'], ['A', 'Z'], ['B', 'Z'], ['C', 'Z']]

for i in data:

    score = 0

    if player_two[i[1]] == 3: #draw
        score += player_one[i[0]][1] + player_two[i[1]]
    
    elif player_two[i[1]] == 0: #lose
        if player_one[i[0]][0] == 'Rock':
            score += player_one['C'][1]
        elif player_one[i[0]][0] == 'Paper':
            score += player_one['A'][1]
        else:
            score += player_one['B'][1]

    elif player_two[i[1]] == 6: #win
        if player_one[i[0]][0] == 'Rock':
            score += player_one['B'][1] + 6
        elif player_one[i[0]][0] == 'Paper':
            score += player_one['C'][1] + 6
        else:
            score += player_one['A'][1] + 6

    scores.append(score)

for i in range(len(scores)):
    print(player_two[data[i][1]],player_one[data[i][0]], data[i], scores[i])

print(sum(scores))
```
