```python
filepath = '../DataFiles/data_day_0002.txt'

with open(filepath, 'r') as fp:

    data = fp.read()
```

```python
data = data.split('\n')

for i in range(len(data)):

    data[i] = data[i].split(' ')

data
```

```python
player_one = {
    'A': 'Rock',
    'B': 'Paper',
    'C': 'Scissors'
}

player_two = {
    'X': ['Rock', 1],
    'Y': ['Paper', 2],
    'Z': ['Scissors', 3]
}

scores = list()

temp_data = [['C', 'X'], ['A', 'Y'], ['B', 'Z']]

for i in data:

    score = 0

    if player_one[i[0]] == player_two[i[1]][0]:
        score += player_two[i[1]][1] + 3
    elif player_one[i[0]] == 'Rock' and player_two[i[1]][0] == 'Paper':
        score += player_two[i[1]][1] + 6
    elif player_one[i[0]] == 'Paper' and player_two[i[1]][0] == 'Scissors':
        score += player_two[i[1]][1] + 6
    elif player_one[i[0]] == 'Scissors' and player_two[i[1]][0] == 'Rock':
        score += player_two[i[1]][1] + 6
    else:
        score += player_two[i[1]][1] + 0

    scores.append(score)

for i in range(len(scores)):
    print(player_two[data[i][1]],player_one[data[i][0]], data[i], scores[i])

print(sum(scores))
```
