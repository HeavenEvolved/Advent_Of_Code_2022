```python
filepath = '../../../Advent_Of_Code_2022/DataFiles/data_day_0008.txt'

with open(filepath, 'r') as fp:

    data = fp.read().strip().split('\n')

test_data = '30373\n25512\n65332\n33549\n35390'.split('\n')
```

```python
visible = [[0 for j in i] for i in data]

for row in range(1, len(data) - 1):
    for col in range(1, len(data[0]) - 1):
        current = int(data[row][col])

        # looking up
        for i in range(row - 1, -1, -1):
            if int(data[i][col]) >= current:
                break
        top = row - i

        # looking down
        for i in range(row + 1, len(data)):
            if int(data[i][col]) >= current:
                break
        bottom = i - row


        # looking right
        for i in range(col + 1, len(data[0])):
            if int(data[row][i]) >= current:
                break
        right = i - col

        # looking left
        for i in range(col - 1, -1, -1):
            if int(data[row][i]) >= current:
                break
        left = col - i

        visible[row][col] = left * right * top * bottom

print(max(map(max, visible)))
```
