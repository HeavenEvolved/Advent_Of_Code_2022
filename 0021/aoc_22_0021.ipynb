```python
monkey_dict = {}

monkey_item_counter = {}

for k in range(monkey_count):
    
    monkey_dict.setdefault(str(k), list())
    monkey_item_counter.setdefault(str(k), 0)

temp_monkey_dict = monkey_dict

print(monkey_dict)

for m in range(20):

    monkey_counter = 0
    print('\n\nRound Starts\n\n')

    for i in range(len(data)):

        operation = data[i]

        if operation.startswith('Starting items:') and m == 0:

            for item in range(len(operation[len('Starting items:'):].strip().replace(':', '').split(', '))):

                temp_monkey_dict[str(monkey_counter)].insert(item, operation[len('Starting items:'):].strip().replace(':', '').split(', ')[item])

        if operation.startswith('Operation: new = old'):

            step = operation[len('Operation: new = old'):].strip().split(' ')
            
            # print(step)

            for j in range(len(temp_monkey_dict[str(monkey_counter)])):

                if step[1] == 'old':

                    value = int(temp_monkey_dict[str(monkey_counter)][j])

                else:

                    value = int(step[1])

                # print(step)

                match step[0]:
                    
                    case '*':   temp_monkey_dict[str(monkey_counter)][j] = str(int(temp_monkey_dict[str(monkey_counter)][j]) * value)

                    case '+':   temp_monkey_dict[str(monkey_counter)][j] = str(int(temp_monkey_dict[str(monkey_counter)][j]) + value)

                temp_monkey_dict[str(monkey_counter)][j] = str(int(temp_monkey_dict[str(monkey_counter)][j]) // 3)

                # print(step, value, monkey_dict[str(monkey_counter)])

                monkey_item_counter[str(monkey_counter)] += 1

            print(monkey_dict)

        if operation.startswith('Test:'):

            curr_monkey_list = temp_monkey_dict[str(monkey_counter)][::-1]

            divisor = int(operation[len('Test:'):].strip().split()[2])

            for j in range(len(temp_monkey_dict[str(monkey_counter)])):

                # print(test_data[i+1][len('If true:'):].strip().split()[-1], int(monkey_dict[str(monkey_counter)][j]) % divisor == 0, divisor, monkey_dict[str(monkey_counter)][j], test_data[i+2][len('If false:'):].strip().split()[-1])

                if int(curr_monkey_list[-1]) % divisor == 0:

                    new_loc = (data[i+1][len('If true:'):].strip().split())[-1]
                
                else:

                    new_loc = (data[i+2][len('If false:'):].strip().split()[-1])[-1]

                temp_monkey_dict[new_loc].append(curr_monkey_list[-1])

                curr_monkey_list.pop()

                temp_monkey_dict[str(monkey_counter)] = curr_monkey_list

                print(new_loc, monkey_dict[new_loc])

        if operation.startswith('If false: ') and monkey_counter < monkey_count:

            monkey_counter += 1 

        # print(monkey_dict, operation)
    monkey_dict = temp_monkey_dict
    print(monkey_dict, monkey_item_counter)
    print('\n\nRound Ends\n\n')
    # print(monkey_dict)

result = sorted(monkey_item_counter, key=monkey_item_counter.get, reverse=True)[:2]

print(monkey_item_counter[result[0]] * monkey_item_counter[result[1]])
```
